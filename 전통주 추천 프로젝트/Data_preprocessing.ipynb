### 정규표현식 패키지 re를 활용

```python
import re
```

### 구매처 텍스트 분리 
- 기존 데이터에서 평점이 포함되어있는 앞 세 글자 제거
- 뒤의 리뷰 작성일도 제거하기 위해 정규표현식을 활용해 숫자 제거

```python
pattern = "[0-9]"
split_list = []
for i in range(0, len(df)):
    a = re.split(pattern, df['date'][i][3:]) # 
    split_list.append(a[0])
```

```python
split_list[:3]
```

### 위의 텍스트에서 이제 구매처만 분리하자
- 구매처 중에선 대문자 영어를 포함한 경우가 있다.
- 때문에 숫자와 소문자 영어를 제거하면 구매처 텍스트를 얻을 수 있을 것이다.

```python
buy = []
pattern = "[0-9a-z]"
for i in range(0, len(split_list)):
    a = re.split(pattern, split_list[i])
    buy.append(a[0])
```

```python
df['buy'] = buy
```

```python
buy[:3] # 잘 분리되었다!!
```

### 리뷰 작성일 추출
- 평점을 제거한 텍스트를 활용
- 등록일은 yy.mm.dd의 형식이다.
- 형식에 맞춰 정규표현식 패턴 지정하면 해당 패턴에 맞는 텍스트만 추출할 수 있을 것이다.

```python
r_date = []
pattern = "([0-9]+.)\s*([0-9]+.)\s*([0-9]+.)"
for i in range(0, len(df)):
    a = re.search(pattern, df['date'][i])
    r_date.append(a.group())
    
df['real_date'] = r_date
```

```python
r_date[:3] ## 완벽~
```
